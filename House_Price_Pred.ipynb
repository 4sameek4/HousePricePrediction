```python
#Importing the required packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#Reading the dataset
data = pd.read_csv('C:/Users/bhatt/OneDrive/Desktop/Sameek/0. University of North Texas/UNT Masters Course/Semester 1 - Fall 2024/CSCE 5214 - Software Development for AI/Project 2/Data/train.csv')
```

```python
df = data
```

```python
# Calculate the correlation matrix
correlation_matrix = df.corr()
```

```python
# Set up the matplotlib figure
plt.figure(figsize=(25, 18))

# Draw the heatmap
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True)

# Add title
plt.title('Correlation Matrix of Features')

# Show the plot
plt.show()
```

```python
# Choose the feature to correlate with others
target_feature = 'SalePrice'
```

```python
# Calculate correlation of the target feature with all others
correlation_with_target = df.corr()[target_feature].drop(target_feature)
```

```python
# Plot the correlation
plt.figure(figsize=(16, 12))
sns.barplot(x=correlation_with_target.index, y=correlation_with_target.values, palette='coolwarm')
plt.xticks(rotation=45)
plt.title(f'Correlation of {target_feature} with Other Features')
plt.xlabel('Features')
plt.ylabel('Correlation Coefficient')
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')  # Line at y=0 for reference
plt.tight_layout()
plt.show()
```
